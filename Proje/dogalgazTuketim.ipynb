```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### <span style="color:pink">Kullanıcı Toplam değerlerini görüntüleme</span>

```python
plt.figure(figsize=(25,7))
sns.histplot( x="Kullanıcı Sınıfı" ,data=df2)
plt.xticks(rotation=90);
```

### <span style="color:pink">İlçelere göre doğalgaz tüketim </span>

```python
dogalgazTüketim = df2.groupby(['İlçe','Yıl'])['Düzeltilmiş Tüketim (m3)'].sum().reset_index()
```

```python
dogalgazTüketim
```

```python
dogalgazTüketim.max()
```

```python
plt.figure(figsize=(18,5))
sns.barplot(x = "İlçe", y = "Düzeltilmiş Tüketim (m3)", data = dogalgazTüketim)
plt.xticks(rotation=90);
```

### <span style="color:pink">Yıllara göre kullanıcı sınıflarının ortalama doğalgaz tüketim verisi </span>

```python

category1 = df2.groupby(["Yıl","Kullanıcı Sınıfı"])[["Düzeltilmiş Tüketim (m3)"]].mean().reset_index()
category1
```

```python
category1.min()
```

```python
plt.figure(figsize=(19,5))
sns.lineplot( x = "Kullanıcı Sınıfı", y = "Düzeltilmiş Tüketim (m3)",hue = "Yıl",data = category1)
plt.xticks(rotation=90);
```

### <span style="color:pink">Kullanıcı sınıfların toplam sayaç bilgisi </span>
