```python
df = data.drop("Unnamed: 0",axis=1)  #axis=1 sütunlar üzerinde işlem yapılcağını belirtir. unnamed:0 sütunu silme işlemi
df.head()
```

```python
df = df.rename(columns={'Unnamed: 1': 'Retailer', 'Unnamed: 2':'Retailer ID', 'Unnamed: 3':'Invoice Date', 'Unnamed: 4':'Region', 'Unnamed: 5':'State',
       'Unnamed: 6':'City', 'Unnamed: 7':'Product', 'Unnamed: 8':'Price per Unit', 'Unnamed: 9':'Units Sold', 'Unnamed: 10':'Total Sales',
       'Unnamed: 11':'Operating Profit', 'Unnamed: 12':'Operating Margin', 'Unnamed: 13':'Sales Method'}) #yeniden adlandırma
```

```python
from plotly.subplots import make_subplots 
import plotly.graph_objects as go
```

```python
# ploty kütüphanesinin alt modüllerini içeri aktarırlar
# suplots alt grafik oluşturmak için kullanılır 
# graph_objects ise grafik nesneleri oluşturmak için kullanılır. grafik nesnelerini (örneğin, çizgi grafikleri, çubuk grafikleri, dağılım grafikleri vb.) içerir.
```

```python
date_method_sold = df.groupby(["Invoice Date","Sales Method"])[["Units Sold"]].sum().reset_index()
date_method_sold
```

```python
in_store = date_method_sold[date_method_sold["Sales Method"]=="In-store"]
online = date_method_sold[date_method_sold["Sales Method"]=="Online"]
outlet = date_method_sold[date_method_sold["Sales Method"]=="Outlet"]
```

```python
in_store
```

```python
fig = make_subplots(rows = 3, cols= 1) # üç satır ve bir sütündan (2*1) oluşan alt grafik oluştur
#append_trace mevcüt bir figüre yeni bir nesne grafiği oluşturulmasını sağlar (bunun yerine add_trace kullanabilirsin .)
fig.append_trace(go.Line(x = in_store["Invoice Date"], y = in_store["Units Sold"]), row = 1, col = 1)
# birinci satırı instore da tutulan  date bilgisini x eksenine y eksenine ise satış bilgisini  
fig.append_trace(go.Line(x = online["Invoice Date"], y = online["Units Sold"]), row = 2, col = 1)
#go.line ile çizgi grafiği olucağını belirtirsin
fig.append_trace(go.Line(x = outlet["Invoice Date"], y = outlet["Units Sold"]), row = 3, col = 1)

fig.show()
```

```python
fig = make_subplots(x_title = "Invoice Date",y_title = "Units Sold",rows = 3, cols= 1)

fig.append_trace(go.Line(x = in_store["Invoice Date"], y = in_store["Units Sold"], name="In-store"), row = 1, col = 1)

fig.append_trace(go.Line(x = online["Invoice Date"], y = online["Units Sold"], name="Online"), row = 2, col = 1)

fig.append_trace(go.Line(x = outlet["Invoice Date"], y = outlet["Units Sold"], name="Outlet"), row = 3, col = 1)

fig.update_yaxes(range=[0,13000], row=3, col=1)
#update_yaxes ile y ekseni üzerinde 3.satırda ki alt grafikte range=[0,13000] kullanımla gösterim aralığının 0-13000 arasında olucağını belirtir
fig.update_layout(title_text = "Units Sold By Invoice Date")
#update_layout : Bu fonksiyon, grafik düzenini ve çeşitli düzenleme seçeneklerini güncellemek için kullanılır.
fig.show()
```
